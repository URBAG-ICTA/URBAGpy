```python
GeoMetData = pd.read_csv('./DataGeo.csv',parse_dates=['DATA'])
```

```python
GeoMetData['diaSet'] = int
GeoMetData['diaMes'] = int
GeoMetData['mes'] = int
GeoMetData['any'] = int
GeoMetData['hora'] = int
```

```python
GeoMetData
```

```python
GeoMetData['diaSet'] = GeoMetData.apply( lambda row: row.DATA.weekday() ,axis=1)
GeoMetData['diaMes'] = GeoMetData.apply( lambda row: row.DATA.day ,axis=1)
GeoMetData['mes'] = GeoMetData.apply( lambda row: row.DATA.month ,axis=1)
GeoMetData['any'] = GeoMetData.apply( lambda row: row.DATA.year ,axis=1)
```

```python
GeoMetData
```

```python
GeoMetData['hour'] = GeoMetData.apply( lambda row: row.DATA.hour ,axis=1)
GeoMetData['minute'] = GeoMetData.apply( lambda row: row.DATA.minute ,axis=1)
GeoMetData['time'] = GeoMetData.apply( lambda row: row.DATA.time() ,axis=1)
```

```python
GeoMetData
```

```python
GeoMetData = GeoMetData.drop(['hora'], axis=1)
```

```python
GeoMetData.to_csv('./DataGeoTime.csv', index=False)
```
