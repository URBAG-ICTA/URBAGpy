```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,10,20)
y = x**2
x,y
```

```python
# We can define figure scale and resolution with figSize and dpi
figure = plt.figure(figsize=(8,4), dpi = 100)
axis = figure.add_axes([0.1,0.1,0.8,0.9])
axis.plot(x,y,'purple')
```

```python
figure.savefig('NicePlot.png')
```

```python
fig = plt.figure(figsize=(8,6), dpi=50)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3, 'red')
```

```python
fig = plt.figure(figsize=(4,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, 'blue', label = "square")
ax.plot(x,x**3, 'red', label = "cube")
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('quin gràfic')
ax.legend()
```

```python
fig = plt.figure(figsize=(4,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, 'blue', label = "square", lw=10)
ax.plot(x,x**3, 'red', label = "cube", lw=1, ls="--")
ax.plot(x, 100*x, 'orange', label = "linear", lw=1, ls='', marker="P")
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('quin gràfic')
ax.legend()
```

```python
#Set plot range
fig = plt.figure(figsize=(4,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, 'blue', label = "square", lw=10)
ax.plot(x,x**3, 'red', label = "cube", lw=1, ls="--")
ax.plot(x, 100*x, 'orange', label = "linear", lw=1, ls='', marker="P")
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('quin gràfic')
ax.legend()
#Set plot range
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
