```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

<h3> Functional aproach to create plots</h3>

```python
# datapoints 20 datapoints from 0 to 10
x = np.linspace(0,10,20)
y = x**2
x,y
```

```python
# simple plot
plt.plot(x,y)
plt.show()
```

```python
# Add title and axis labels to the plot
plt.plot(x,y)
plt.title('Nice plot')
plt.xlabel('X horizontal')
plt.ylabel('Y vertical')
plt.show()
```

```python
# plot several plots in the same canvas
# .subplot(nrows,ncols,plotIndex)

plt.subplot(2,3,1)
plt.plot(x,y, 'red')

plt.subplot(2,3,2)
plt.plot(x,y, 'green')

plt.subplot(2,3,3)
plt.plot(x,y, 'blue')

plt.subplot(2,3,4)
plt.plot(x,y, 'yellow')

plt.subplot(2,3,5)
plt.plot(x,y, 'black')

plt.subplot(2,3,6)
plt.plot(x,y, 'orange')

plt.show()
```

<h3> Object oriented aproach to create plots</h3>

```python
figure = plt.figure()
axis = figure.add_axes([0,0,0.8,0.9])
axis.plot(x,y,'purple')
```

```python
figure = plt.figure()
axis = figure.add_axes([0,0,1,1]) # left. bottom, width and height
axis.plot(x,y,'purple')
```

```python
figure = plt.figure()
ax = figure.add_axes([0,0,1,1]) # left. bottom, width and height
ax.plot(x,y,'purple')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('quin gràfic')
```

```python
fig =plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([.1,0.4,0.4,0.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig =plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([1.2,0,1,1])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig =plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([1.2,0,1,1])

axes1.plot(x,y,'purple')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('quin gràfic')

axes2.plot(x,y,'green')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('awesome')
```
